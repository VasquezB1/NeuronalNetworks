```python
# Funciones para guardar y cargar un Pipeline

def guardarPipeline(pipeline,nombreArchivo):
    with open(nombreArchivo+'.pickle', 'wb') as handle:
        pickle.dump(pipeline, handle, protocol=pickle.HIGHEST_PROTOCOL)


def cargarPipeline(nombreArchivo):
    with open(nombreArchivo+'.pickle', 'rb') as handle:
        pipeline = pickle.load(handle)
    return pipeline
```

```python
nombreArchivoPreprocesador='pipePreprocesadores'
pipe=cargarPipeline(nombreArchivoPreprocesador)
pipe.steps
```

Creación de Modelo Base y Evaluación con cross_val_score

```python

train=copy.deepcopy(dataframe)
test=copy.deepcopy(Y)
train=train.fillna(0)
test=test.fillna(0)
```

```python
def build_model():
    model = Sequential()
    model.add(Dense(700, input_shape=(cantidadVariablesIndependientes,), activation='relu'))
    model.add(Dense(500, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
    return model
```

```python
#El modelo se pasa como parámetro
estimator = KerasClassifier(build_fn=build_model, epochs=10000, batch_size=64)
pipe.steps.append(['modelNN',estimator])


#numFolds es la cantidad de Folds (conjuntos) para el entrenamiento del modelo
numFolds=5
#kfold es la configuración de validación K-Fold Cross-Validation
kfold = ShuffleSplit(n_splits=numFolds)
```

```python
#n_jobs es para ocupar mas de un procesador. El parámetro -1 indica que se utilizarán todos los procesadores disponibles
cvs=cross_val_score(pipe, train, test, cv=kfold, n_jobs=-1)
```

## Optimización y Fine Tuning con GridSearchCV: Compilación

```python
def GetTime(gs_time):
    sec=timedelta(seconds=gs_time)
    d = datetime(1,1,1) + sec
    tiempoTotal=("%d días: %d horas: %d min: %d seg" % (d.day-1, d.hour, d.minute, d.second))
    return tiempoTotal
```

```python

pipe=cargarPipeline(nombreArchivoPreprocesador)
pipe.fit_transform(dataframePreprocesado)

train=copy.deepcopy(dataframe)
test=copy.deepcopy(Y)

def build_model(optimizer):
    model = Sequential()
    model.add(Dense(10, input_shape=(cantidadVariablesIndependientes,), activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(8, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['acc'])
    return model


batch_size = [1,5,10,15,20,25,30,40,50,60,70,80,90,100]  #14
epochs = [2,4,6,8,10,15,20,25,30,40,50] #11
optimizer = ['RMSprop','SGD','Adam'] #3


#parámetros que queremos probar, y sus valores 
parameters = {'modelNN__batch_size': batch_size,
             'modelNN__epochs': epochs,
             'modelNN__optimizer': optimizer}


estimator = KerasClassifier(build_fn=build_model, verbose=0)


pipe.steps.append(['modelNN',estimator])


#Ahora no le pasamos los parámetros al KerasClasifier, porque los pasaremos a través de GridSearchCV
#el argumento verbose=0 es para que no muestre salida, si lo dejamos en cero, no mostrará la barra de progreso del entrenamiento
#GridSearchCV: recibe como parámetros el modelo, los parámetros, la medida sobre la que queremos comparar, y la 
#cantidad de veces que lo entrenará para sacar la media de accuracy.
tic = time()


grid_search = GridSearchCV(estimator=pipe, param_grid=parameters, scoring='accuracy', cv=5, n_jobs=-1)


grid_result = grid_search.fit(train, test)


print(grid_result.best_params_)


#Para revisar todos los parámetros existentes en el Pipeline
#pipe.get_params().keys()


#Algunos resultados:
#Best: 0.764000 using {'modelNN__batch_size': 70, 'modelNN__epochs': 30, 'modelNN__optimizer': 'RMSprop'}
#Best: 0.769000 using {'modelNN__batch_size': 70, 'modelNN__epochs': 30, 'modelNN__optimizer': 'Adam'}


gs_time = time() - tic
print('Tiempo en segundos: ',gs_time)
tiempoTotal=GetTime(gs_time)
print('Tiempo: ',tiempoTotal)


gs1=grid_result
```

```python
print("Best: %f using %s" % (gs1.best_score_, gs1.best_params_))
```

```python
means = gs1.cv_results_['mean_test_score']
stds = gs1.cv_results_['std_test_score']
params = gs1.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

# Optimización y Fine Tuning con GridSearchCV: Densidad de las Capas de Neuronas

```python
pipe=cargarPipeline(nombreArchivoPreprocesador)
pipe.fit_transform(dataframePreprocesado)

train=copy.deepcopy(dataframe)
test=copy.deepcopy(Y)


#Notemos que se incluyen los mejores parámetros del paso de optimización anterior (batch_size, epochs y optimizer)
def build_model(l1,l2,l3):
    model = Sequential()
    model.add(Dense(l1, input_shape=(cantidadVariablesIndependientes,), activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(l2, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(l3, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['acc'])
    return model


l1=[2,4,8,16,32,64,96,128,160,192,224,256,512]
l2=[2,4,8,16,32,64,96,128,160,192,224,256,512]
l3=[2,4,8,16,32,64,96,128,160,192,224,256,512]
parameters = parameters = {'modelNN__l1':l1,
                           'modelNN__l2':l2,
                           'modelNN__l3':l3}


estimator = KerasClassifier(build_fn=build_model, verbose=0, batch_size=70, epochs=30)


pipe.steps.append(['modelNN',estimator])


tic = time()


grid_search = GridSearchCV(estimator=pipe, param_grid=parameters, scoring='accuracy', cv=5, n_jobs=-1)
grid_result = grid_search.fit(train, test)


#Algunos resultados:
#Adam: Best: 0.765000 using {'modelNN__l1': 16, 'modelNN__l2': 32, 'modelNN__l3': 192}
#Rmsprop: Best: 0.770000 using {'modelNN__l1': 16, 'modelNN__l2': 128, 'modelNN__l3': 256}
#Adam: Best: 0.771000 using {'modelNN__l1': 64, 'modelNN__l2': 2, 'modelNN__l3': 64}


gs_time1 = time() - tic
print('Tiempo en segundos: ',gs_time1)
tiempoTotal=GetTime(gs_time1)
print('Tiempo: ',tiempoTotal)


gs2=grid_result
```

# Optimización y Fine Tuning con GridSearchCV: Regularización de Dropout

```python
pipe=cargarPipeline(nombreArchivoPreprocesador)
pipe.fit_transform(dataframePreprocesado)

train=copy.deepcopy(dataframe)
test=copy.deepcopy(Y)


#Notemos que se incluyen los mejores parámetros de los pasos de optimización anteriores (batch_size, epochs, optimizer, 
#cantidad de neuronas en la capa 1, capa 2 y capa 3).
def build_model(d1, d2, d3):
    model = Sequential()
    model.add(Dense(64, input_shape=(cantidadVariablesIndependientes,), activation='relu'))
    model.add(Dropout(d1))
    model.add(Dense(2, activation='relu'))
    model.add(Dropout(d2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(d3))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['acc'])
    return model


d1=[0,0.05,0.1,0.15,0.2,0.3,0.4,0.5]
d2=[0,0.05,0.1,0.15,0.2,0.3,0.4,0.5]
d3=[0,0.05,0.1,0.15,0.2,0.3,0.4,0.5]
parameters = parameters = {'modelNN__d1':d1,
                           'modelNN__d2':d2,
                           'modelNN__d3':d3}


estimator = KerasClassifier(build_fn=build_model, verbose=0, batch_size=70, epochs=30)


pipe.steps.append(['modelNN',estimator])


tic = time()


grid_search = GridSearchCV(estimator=pipe, param_grid=parameters, scoring='accuracy', cv=5,n_jobs=-1)
grid_result = grid_search.fit(train, test)


#Algunos resultados:
#Adam: Best: 0.763000 using {'modelNN__d1': 0.2, 'modelNN__d2': 0.5, 'modelNN__d3': 0.2}
#Rmsprop: Best: 0.767000 using {'modelNN__d1': 0.2, 'modelNN__d2': 0.5, 'modelNN__d3': 0.2}
#Adam: Best: 0.763000 using {'modelNN__d1': 0.2, 'modelNN__d2': 0.05, 'modelNN__d3': 0}


gs_time3 = time() - tic
print('Tiempo en segundos: ',gs_time3)
tiempoTotal=GetTime(gs_time3)
print('Tiempo: ',tiempoTotal)


gs3=grid_result
```

# Predicción con Modelo Optimizado y Evaluación con métricas de calidad: MAE, MSE, RMSE, Accuracy, Precision, Recall y F1

```python
y_pred = (pipe.predict(X_test) > 0.5).astype("int32")
y_pred=y_pred.flatten()# de 2D a 1D

dataframeFinal=pd.DataFrame({'real':y_test, 'predicción': y_pred})


np.set_printoptions(formatter={'float': lambda X: "{0:0.0f}".format(X)})


dataframeFinal.head(10)
```

```python
MAE=metrics.mean_absolute_error(y_test, y_pred)
MSE=metrics.mean_squared_error(y_test, y_pred)
RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
accuracy=str(accuracy_score(y_test, y_pred))

MAE=str(round(MAE, 4))
MSE=str(round(MSE, 4))
RMSE=str(round(RMSE, 4))


print('Mean Absolute Error (MAE):', MAE)
print('Mean Squared Error (MSE):', MSE)
print('Root Mean Squared Error (RMSE):', RMSE) 
print('Accuracy: ' + accuracy) 


print('Confusion_matrix:')
y_test_transformado = y_test
y_pred_transformado = y_pred


cm = confusion_matrix(y_test_transformado, y_pred_transformado)
print(cm)
tn, fp, fn, tp = confusion_matrix(y_test_transformado, y_pred_transformado).ravel()
print(tn,fp,fn,tp)


precision=precision_score(y_test_transformado, y_pred_transformado)
precision=str(round(precision, 4))
print('Precision: ',precision)


recall=recall_score(y_test_transformado, y_pred_transformado)
recall=str(round(recall, 4))
print('Recall: ',recall)


f1=f1_score(y_test_transformado, y_pred_transformado)
f1=str(round(f1, 4))
print('F1:',f1)
```
